# Infrastructure Spending

```python
import pandas as pd

pd.set_option('display.max_colwidth', 500)
pd.set_option('display.max_rows', 500)
pd.options.display.float_format = '${:,.2f}'.format
```

## Load Data

```python
infrastructure_url = 'http://infrastructure.gc.ca/alt-format/opendata/project-list-with-forcast-dates-liste-de-projets-avec-dates-prevu-en.csv'
projects = pd.read_csv('../data/project-list-en.csv', encoding='ISO-8859-1').set_index('projectNumber')
```

```python
open_cond = ~(projects.constructionStartDate.isnull()) & (projects.constructionEndDate.isnull())
closed_cond = ~(projects.constructionEndDate.isnull())
```

## Closed Projects

```python
closed_projects = projects[closed_cond]
closed_projects
```

## Open Projects

```python
open_projects = projects[open_cond]
open_projects
```

## Locations

```python
projects.groupby('location').sum().sort_values('federalContribution', ascending=False)
```
